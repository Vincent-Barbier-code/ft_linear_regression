Import des bibliotheques, chargement du csv et initialisation des donnees.
θ0 = m 
θ1 = p

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open('data.csv', 'r') as file:
	data = csv.reader(file)
	next(data)

	km = []
	price = []
 
	for row in data:
		# print(row['km'], row['price'])
		km.append(float(row[0]))
		price.append(float(row[1]))
	
m = 0
p = 0
```

Nuage de points pour prouver l utilisation de la fonction lineaire simple.

```python
plt.title("Prix en fonction du kilometrage")
plt.plot(km, price, "ob") # ob = type de points "o" ronds, "b" bleus

plt.show() 
```

Fonction pour avoir le coefficient directeur de la fonction affine.

estimatePrice(km) = (θ1 ∗ km) + θ0 

θ0 (p) = ordonnee a l origine
θ1 (m) = coefficient directeur

ce qui revient a :
f(x) = (x * m) + p	
f(x1) = y1				ou y1 = x1 * m + p
Donc pour 2 points sur notre fonction affine : f(x1) = y1 et f(x2) = y2

Maintenant on isole p:
y1 - p = (x1 * m)
-p = (x1 * m) - y1

p = - (x1 * m) + y1

de la meme facon pour l'autre point p = - (x2 * m) + y2.

Donc:
-(x1 * m) + y1 = -(x2 * m) + y2
-(x1 * m) + y1 + (x2 * m) - y2 = 0
m * (x2 - x1) + y1 - y2 = 0
m * (x2 - x1) = y2 - y1
m = (y2 - y1) / (x2 - x1)


```python
def coefficient_directeur(x1, x2, y1, y2): # notre m
	deltaX = x2 - x1
	deltaY = y2 - y1
	return deltaY / deltaX
m = coefficient_directeur(km[0], km[18], price[0], price[18]) #θ0
```

Fonction pour avoir la coordonne a l origine de la fonction affine.
Comme on a vu avant p = - (x1 * m) + y1

```python
def ordonne_origine(x1 , y1, m):
	return - m * x1 + y1

p = ordonne_origine(km[0], price[0], m) #θ1
print(m, p)
```

Fonction pour avoir l estimation du prix par rapport au km

```python
def estimatePrice(x, m, p):
    return m * x + p


# Dans notre cas :
Kilometers = 100000
PredictedPrice = estimatePrice(Kilometers, m, p)
print("Pour", Kilometers, "km, le prix estimé est de", PredictedPrice, "€")
```

Estimation du prix affiche sur le graphe.

```python
plt.title("Prix en fonction du kilometrage")
plt.plot(km, price, "ob") # ob = type de points "o" ronds, "b" bleus
plt.plot(Kilometers, PredictedPrice, "or") # or = type de points "o" ronds, "r" rouges
plt.show() 
```

Deuxieme programme de fonction lineaire  regressive.
Affichage de la regression lineraire.
Reinitialisation des variables.

```python
plt.title("Prix en fonction du kilometrage")
plt.plot(km, price, "ob") # ob = type de points "o" ronds, "b" bleus
plt.plot(Kilometers, PredictedPrice, "or") # or = type de points "o" ronds, "r" rouges
plt.plot([km[0], km[18]], [estimatePrice(km[0], m, p), estimatePrice(km[18], m, p)], "g-") 
plt.show()

m = 0
p = 0
```
